```python
for xxx ,xxxml in enumerate (sub_files):
    #print(xml)
    shutil.copy( results_path+'/'+str(xxxml), 'try/img'+str(xxx+Img_number)+'.xml')
    updateTree = ET.parse('try/img'+str(xxx+Img_number)+'.xml')   # 读取待修改文件
    root = updateTree.getroot()
    sub2 =root.find('filename')
    sub2.text = 'img'+str(xxx+Img_number)+'.jpg'
    updateTree.write('try/img'+str(xxx+Img_number)+'.xml',encoding='utf-8')
```
