```python
%matplotlib inline
import cv2
import cv2 as cv
import numpy as np
import imutils
import matplotlib.pyplot as plt
from skimage import data,color,morphology
```

```python
img1 = cv.imread('try2.jpg',0)
img2=img1.copy()
```

```python
img3 = img2.copy()
cv2.threshold(img2,80,255,0,img2)
for i in range(0,img2.shape[0]):
    for j in range(0,img2.shape[1]):
        img3[i,j] = 255-img2[i,j]
```

```python
plt.imshow(img2,cmap='gray'),plt.imshow(img3,cmap='gray')
```

```python
(_, cnts, _)= cv2.findContours(img3, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
for i in range(len(cnts)):
    area = cv2.contourArea(cnts[i])
    if area < 900:
        cv2.drawContours(img3,[cnts[i]],0,0,-1)
img5 = img3.copy()
plt.imshow(img5)
```

```python
img5=img5.copy()
(contours, cnts1, _) = cv2.findContours(img5, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
for j in range(len(cnts1)):
    area1 = cv2.contourArea(cnts1[j])
    print(area1)
    if area1 ==157.0:
        cv2.drawContours(img5,[cnts1[j]],0,0,-1)
    elif area1==261.5:
        cv2.drawContours(img5,[cnts1[j]],0,0,-1)
    elif area1==568.0:
        cv2.drawContours(img5,[cnts1[j]],0,0,-1)
```
